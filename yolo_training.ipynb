###Model Initialization

###Starting training

````python
results = model.train(
    data='data.yaml',
    epochs = 50,
    save=True,
    device = 'gpu',
    imgsz=640,
    save_period=1,
    project='/content/drive/MyDrive/yolov8_weights',
    name='project'
)
````
